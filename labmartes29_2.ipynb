```python
import matplotlib.pyplot as plt

# Setting up a demonstration of the Central Limit Theorem (CLT) using daily sales data
fig, ax = plt.subplots(1, 2, sharey=True, squeeze=True, figsize=(11, 5))

# Assume sales amounts are uniformly distributed between $1000 and $5000
lower_bound_A, upper_bound_A = 10, 15
lower_bound_B, upper_bound_B = 10, 60
sample_sizes = [100, 1000,]  # Different sample sizes to demonstrate the CLT

for i, n in enumerate(sample_sizes):
    # Generating random daily sales data for 'n' days
    daily_sales_A = np.random.uniform(lower_bound_A, upper_bound_A, (100, n))

for i, n in enumerate(sample_sizes):
    # Generating random daily sales data for 'n' days
    daily_sales_B = np.random.uniform(lower_bound_B, upper_bound_B, (1000, n))

  # Calculating the mean of daily sales for each of the 10,000 samples
    sample_means_A = np.mean(daily_sales_A, axis=1)

  # Calculating the mean of daily sales for each of the 10,000 samples
    sample_means_B = np.mean(daily_sales_B, axis=1)

    # Plotting the histogram of the sample means
    ax[i].hist(sample_means_A, bins=10, density=True, color=[0, 0.2, .8, .6])
    ax[i].set_title(f'n={n}')

    # Plotting the histogram of the sample means
    ax[i].hist(sample_means_B, bins=10, density=True, color=[0, 0.2, .8, .6])
    ax[i].set_title(f'n={n}')

    # Fitting a normal distribution to the sample means
    mean, std = np.mean(sample_means_A), np.std(sample_means_A, ddof=1)
    x = np.linspace(min(sample_means_A), max(sample_means_A), 100)
    normal_pdf_A = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-(x - mean) ** 2 / (2 * std ** 2))
    
    # Plotting the fitted normal distribution over the histogram
    ax[i].plot(x, normal_pdf_A, color=[1, 0, 0, .6], linewidth=2)

    # Fitting a normal distribution to the sample means
    mean, std = np.mean(sample_means_B), np.std(sample_means_B, ddof=1)
    x = np.linspace(min(sample_means_B), max(sample_means_B), 100)
    normal_pdf_B = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-(x - mean) ** 2 / (2 * std ** 2))
    
    # Plotting the fitted normal distribution over the histogram
    ax[i].plot(x, normal_pdf_B, color=[1, 0, 0, .6], linewidth=2)

plt.show()
```

El resultado debe ser como el siguiente:



```python
# your code here
#1 . 
import numpy as np
from scipy.stats import uniform

def generar_numeros_aleatorios_scipy(bottom, ceiling, count):
    """
    Genera números aleatorios uniformemente distribuidos entre dos límites usando SciPy.

    Args:
        bottom (float): Límite inferior del rango.
        ceiling (float): Límite superior del rango.
        count (int): Número de valores aleatorios a generar.

    Returns:
        numpy.ndarray: Un array de NumPy con los números aleatorios generados.

    Raises:
        ValueError: Si el límite inferior es mayor o igual al límite superior, o si el número de valores a generar es menor o igual a cero.
    """

    if bottom >= ceiling:
        raise ValueError("El límite inferior debe ser menor que el límite superior.")
    if count <= 0:
        raise ValueError("El número de valores a generar debe ser positivo.")

    return uniform.rvs(loc=bottom, scale=ceiling - bottom, size=count)
```

```python
# your code here
# Primer conjunto de parámetros
numeros_1 = generar_numeros_aleatorios_scipy(bottom=10, ceiling=15, count=100)
print("Primer conjunto de números:\n", numeros_1)

# Segundo conjunto de parámetros
numeros_2 = generar_numeros_aleatorios_scipy(bottom=10, ceiling=60, count=1000)
print("\nSegundo conjunto de números:\n", numeros_2)
```

## Distribución Exponencial

1. Utilizando `numpy.random.exponential`, crea una función que devuelva una lista de números distribuidos exponencialmente con la media de 10. 

1. Utiliza la función para generar dos secuencias de números de tamaño 10 y 100.

1. 2. Traza las distribuciones como histogramas con el número de bins como 100.

El resultado debe ser como el siguiente:



```python
# your code here
def generar_numeros_exponenciales(media, tamaño):
    escala = media  # En la distribución exponencial, la media es igual al parámetro de escala
    return np.random.exponential(scale=escala, size=tamaño)

secuencia_10 = generar_numeros_exponenciales(10, 10)
secuencia_100 = generar_numeros_exponenciales(10, 100)

fig, ax = plt.subplots(1, 2, figsize=(14, 6))


ax[0].hist(secuencia_10, bins=100, alpha=0.7, color='blue')
ax[0].set_title('Histograma de secuencia de tamaño 10')
ax[0].set_xlabel('Valor')
ax[0].set_ylabel('Frecuencia')
ax[0].set_xlim(0,8)

ax[1].hist(secuencia_100, bins=100, alpha=0.7, color='green')
ax[1].set_title('Histograma de secuencia de tamaño 100')
ax[1].set_xlabel('Valor')
ax[1].set_ylabel('Frecuencia')
ax[1].set_xlim(0,600)

# Mostrar los gráficos
plt.tight_layout()
plt.show()


```

¿En qué se diferencian ambas distribuciones?

La media cambia, por lo que la distribución también cambia.
